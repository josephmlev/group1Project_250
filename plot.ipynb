```python
import matplotlib
%matplotlib inline

import numpy as np
import functions as funs
```

```python
ndim =4
nsamples = 10000

# generate some fake chain from gaussian
temp_chain1 = np.zeros((nsamples,ndim))
temp_chain1[:,0] = np.random.normal(loc=67.36, scale=0.7, size=nsamples) #fake H0
temp_chain1[:,1] = np.random.normal(loc=0.3, scale=0.02, size=nsamples) #fake Omega_m
temp_chain1[:,2] = np.random.normal(loc=0.7, scale=0.05, size=nsamples) #fake Omega_de
temp_chain1[:,3] = np.random.normal(loc=-19, scale=0.25, size=nsamples) #fake M

# generate the second chain by shifting the first chain by 0.05
temp_chain2 = temp_chain1 + 0.05
```

```python
names = ['H0', 'Om', 'Ode','M']
labels = ['H_0 [km/s/Mpc]', '\Omega_m', '\Omega_{de}', 'M']
```

```python
# plot using default labeling and legends
ch1 = funs.make_plot([temp_chain1, temp_chain2])
ch1.plot_2d()
ch1.plot_1d()
```

```python
#plot using customized label and legends
ch2 = funs.make_plot([temp_chain1, temp_chain2], legend=['chain1', 'chain2'], label=labels, name=names)
ch2.plot_2d()
ch2.plot_1d()
```
